````python
credits_clean.drop_duplicates(inplace=True)
````

````python
# Check Missing Values

titles_clean.isnull().sum().sort_values(ascending=False)
````

### Column: seasons

- Most missing values are because the title is a Movie.

- A movie has no seasons.
- These are logical missing values, not data errors.
Do not fill or remove these values.

````python
# Check Missing Values

credits_clean.isnull().sum().sort_values(ascending=False)
````

````python
# Column: age_certification

# Many values are missing. Instead of deleting rows: fill with 'Not Rated'


titles_clean['age_certification'].fillna('Not Rated', inplace=True)
````

````python
# Checking again for missing values in 'age_certification' column
titles_clean['age_certification'].isnull().sum()
````

````python
# Column: 'description'

# Very few values are missing.
# Instead of deleting rows: fill with 'Description Not Available'

titles_clean['description'].fillna('Description Not Available', inplace=True)
````

````python
titles_clean['imdb_score'].isnull().sum()
````

````python
# Column: 'imdb_score'
# Instead of deleting rows: Imputed with the 'Median'

titles_clean['imdb_score'].fillna(
    titles_clean['imdb_score'].median(),
    inplace=True
)
````

# Exploratory Data Analysis (EDA)

## Univariate Analysis

In Univariate Analysis, we analyze one variable at a time.

The purpose is to understand the distribution, frequency, and characteristics of individual variables.

## Chart 1. Distribution of Movies and TV Shows

````python
# Chart 1: Distribution of Movies and TV Shows

plt.figure(figsize=(7,4))

ax = sns.countplot(
    data=titles_clean,
    x='type',
    hue='type',
    palette='Set2',
    legend=False
)

# Display count labels on each bar
for container in ax.containers:
    ax.bar_label(container, fontsize=11)

plt.title('Distribution of Movies and TV Shows', fontsize=15, fontweight='bold')
plt.xlabel('Content Type', fontsize=10, fontweight='bold')
plt.ylabel('Number of Titles', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
````

### Why did I choose this chart?

A Count Plot is the most suitable visualization for categorical variables because it compares the frequency of each category. Since the type column contains only MOVIE and SHOW, this chart provides a clear comparison of the platform's content distribution.



### Insights
- Movies constitute the majority of Amazon Prime Video's content library.
- TV Shows represent a smaller portion of the catalog.
- The platform appears to focus more heavily on movie content.

### Business Impact

- Understanding the distribution of content helps Amazon evaluate whether its library is balanced. If TV Shows are underrepresented, increasing investment in original series could improve long-term user engagement and retention.

### Recommendation
- Continue expanding high-performing movie content while evaluating opportunities to increase investment in TV Shows if user demand supports it.
- Analyze viewer engagement separately to determine whether movies or TV shows drive higher watch time.

# Multivariate Analysis
Multivariate Analysis examines the relationship among three or more variables simultaneously. It helps uncover patterns that cannot be identified by analyzing only one or two variables.

# Chart 18: Correlation Heatmap

````python
# Chart 18: Correlation Heatmap
# Objective:
# Analyze relationships among numerical variables.


numerical_columns = [
    'release_year',
    'runtime',
    'imdb_score',
    'imdb_votes',
    'tmdb_popularity',
    'tmdb_score'
]

corr_matrix = titles_clean[numerical_columns].corr()

plt.figure(figsize=(10,7))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Numerical Features",
          fontsize=15,
          fontweight='bold')

plt.tight_layout()

plt.show()
````

### Why this chart?

A heatmap is the best visualization for showing relationships among multiple numerical variables simultaneously.

### Insights
- imdb_score and tmdb_score: These variables share the strongest positive correlation
- imdb_votes and tmdb_popularity: There is a mild positive correlation of 0.26 between these two metrics.
- runtime and Content Performance: runtime exhibits very weak negative correlations with imdb_score (-0.10) and tmdb_score (-0.10).
- runtime and tmdb_popularity: A negligible negative correlation of -0.03 exists here, indicating no meaningful inverse link between duration and viewer traction.


### Business Impact
